–í –Ω–∞–∑–≤–∞–Ω–∏—è—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ pcp 01, pcp 06, pcp 24 –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –ø—Ä–æ–±–µ–ª. –≠—Ç–æ –Ω–µ –æ—á–µ–Ω—å —É–¥–æ–±–Ω–æ, –µ—Å–ª–∏ –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º –≤—ã –ø–ª–∞–Ω–∏—Ä—É–µ—Ç–µ –æ–±—Ä–∞—â–∞—Ç—å—Å—è –∫ –∫–æ–ª–æ–Ω–∫–∞–º —á–µ—Ä–µ–∑ —Ç–æ—á–∫—É, –±–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –∫–∞–≤—ã—á–µ–∫ –∏ —Å–∫–æ–±–æ—á–µ–∫. 

–ó–∞–º–µ–Ω–∏—Ç–µ –ø—Ä–æ–±–µ–ª –≤ –Ω–∞–∑–≤–∞–Ω–∏—è—Ö –Ω–∞ –∑–Ω–∞–∫ –Ω–∏–∂–Ω–µ–≥–æ –ø–æ–¥—á–µ—Ä–∫–∏–≤–∞–Ω–∏—è. 

```python
# 1-—ã–π –≤–∞—Ä–∏–∞–Ω—Ç taxi = taxi.rename(columns = {'pcp 01':'pcp_01', 'pcp 06': 'pcp_06', 'pcp 24':'pcp_24'})
# 2-–æ–π –≤–∞—Ä–∏–∞–Ω—Ç, —á—Ç–æ–±—ã –Ω–µ –ø—Ä–æ–ø–∏—Å—ã–≤–∞—Ç—å –≤—Ä—É—á–Ω—É—é
def underscore_rename(name):
    return name.replace(' ','_')
taxi = taxi.rename(columns=underscore_rename)
```

```python
taxi.borough.value_counts() # 1-—ã–π –≤–∞—Ä–∏–∞–Ω—Ç
# taxi.query('borough == "Brooklyn"').value_counts() 2-–æ–π –≤–∞—Ä–∏–∞–Ω—Ç
```

```python
taxi.groupby('borough') \
    .agg({'pickups':'sum'}) \
    .idxmax() # –ø–æ–∫–∞–∂–µ—Ç –º–∞–∫—Å–∏–º—É–º
```

‚≠êÔ∏è–ó–∞–¥–∞–Ω–∏–µ —Å–æ –∑–≤—ë–∑–¥–æ—á–∫–æ–π!‚≠êÔ∏è

–ï—â–µ –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–æ–ª–µ–∑–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤ –≤ pandas! –í–æ–∑–º–æ–∂–Ω–æ, –≤ –ø—Ä–µ–¥—ã–¥—É—â–µ–º —Å—Ç–µ–ø–µ –≤—ã –ø—Ä–æ—Å—Ç–æ –ø–æ—Å–º–æ—Ç—Ä–µ–ª–∏ –Ω–∞ –¥–∞–Ω–Ω—ã–µ –∏–ª–∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–ª–∏ –∑–Ω–∞—á–µ–Ω–∏—è. –£—Å–∫–æ—Ä–∏—Ç—å –¥–∞–Ω–Ω—ã–π –ø—Ä–æ—Ü–µ—Å—Å –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º –ø–æ–º–æ–≥—É—Ç –º–µ—Ç–æ–¥—ã idxmin() –∏ idxmax(), –∫–æ—Ç–æ—Ä—ã–µ –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç –∏–Ω–¥–µ–∫—Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ –∏–ª–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è.

idxmin

–°–æ—Ö—Ä–∞–Ω–∏—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏–µ —Ä–∞–π–æ–Ω–∞ —Å –Ω–∞–∏–º–µ–Ω—å—à–∏–º —á–∏—Å–ª–æ–º –ø–æ–µ–∑–¥–æ–∫ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é min_pickups, –ø—Ä–∏–º–µ–Ω–∏–≤ –ø–æ–¥—Ö–æ–¥—è—â–∏–π –º–µ—Ç–æ–¥.

```python
taxi.groupby('borough') \
    .agg({'pickups':'sum'}) \
    .idxmin() # –ø–æ–∫–∞–∂–µ—Ç –º–∏–Ω–∏–º—É–º
```

–î–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–∞–π–æ–Ω–∞ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ —á–∏—Å–ª–æ –ø–æ–µ–∑–¥–æ–∫ –ø–æ –º–µ—Å—è—Ü–∞–º. –û—Ç—Å–æ—Ä—Ç–∏—Ä—É–π—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ —É–±—ã–≤–∞–Ω–∏—é –∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∏–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –≤ pickups_by_mon_bor. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –∏—Ç–æ–≥–æ–≤—ã–π –¥–∞—Ç–∞—Å–µ—Ç –¥–æ–ª–∂–µ–Ω —Å–æ—Å—Ç–æ—è—Ç—å –∏–∑ 3-—Ö –∫–æ–ª–æ–Ω–æ–∫ ‚Äì pickup_month, borough, pickups.

```python
taxi.groupby(['borough', 'pickup_month'], as_index = False) \
    .agg({'pickups':'sum'}).sort_values(by = 'pickups', ascending = False)
```
