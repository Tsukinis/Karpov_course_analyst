Для H1 - вероятность появления белой грани = 0.5 <br>
тогда вероятность при 7 бросках белая грань выпадет ровно 1 =>

```python
from math import factorial

def num_of_successes(n, k):
    return factorial(n)/(factorial(k) * factorial(n - k))
```

```python
C = num_of_successes(7,1)
P = C * 0.5**1 * 0.5**6
P
```

Для H2 - вероятность появления белой грани = 1/6 <br>
тогда вероятность при 7 бросках белая грань выпадет ровно 1 =>

```python
C = num_of_successes(7,1)
P = C * (1/6)**1 * (5/6)**6
P
```

Пусть A - самолет в зоне <br>
Пусть B - самолет не обнаружен <br>
тогда вероятность того, что самолен не будет обнаружен, если он в зоне<br>
P(B|A) = 1 - 0.98 = 0.02<br>
А вероятность того, что самолет не обнаружат, если его реально нет в зоне<br>
P(-B|-A) = 1 - 0.05 = 0.95<br>

```python
# P(A|B) = P(B|A) * P(A) / P(B|A) * P(A) + P(-B|-A) * P(-A)
(0.02*0.07)/(0.02*0.07 + 0.95*0.93)
```
