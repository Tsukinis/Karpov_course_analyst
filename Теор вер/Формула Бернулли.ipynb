```python
from math import factorial

def num_of_successes(n, k):
    return factorial(n)/(factorial(k) * factorial(n - k))
```

```python
C = num_of_successes(5,3) # p = 0,4 - отказ, а число испытаний 5
C
```

```python
round(C * 0.4**3 * 0.6**2,4) #откажут 3 эл-та
```

```python
C = num_of_successes(5,0) # p = 0,4 - отказ, а число испытаний 5
C
```

```python
round(1 - C * 0.4**0 * 0.6**5,3) # хотя бы 1 откажет, т.е. 1 - никто не откажет
```
